```python
import gmaps
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Google developer API key
from config import gkey

# Access maps with unique API key
gmaps.configure(api_key=gkey)

csv_path = "../Project-1/EVstation_data.csv"

EVstation_data =pd.read_csv(csv_path)
EVstation_data

```

```python
lng = []
lat = []
coordinates = []

lng = EVstation_data["longitude"]
lat = EVstation_data["latitude"]

for i in range(len(lng)):
    
    coordinates.append([lat[i], lng[i]])
    
    print(coordinates)
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '1200px',
    'height': '1200px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Assign the marker layer to a variable
EV_locations = coordinates

markers = gmaps.marker_layer(EV_locations)


# Add the layer to the map
fig.add_layer(markers)
fig
```
